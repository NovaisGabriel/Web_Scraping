```python
from bs4 import BeautifulSoup as bf4
from soupselect.soupselect import *
from time import sleep
import urllib.request
import numpy as np
import pandas as pd
import ast
import json as js
import re
import random
%load_ext autotime
```

```python
def find_substring(substring, string):
    indices = []
    index = -1  # Begin at -1 so index + 1 is 0
    while True:
        # Find next index of substring, by starting search from index + 1
        index = string.find(substring, index + 1)
        if index == -1:  
            break  # All occurrences have been found
        indices.append(index)
    return indices
```

# Pegando os nomes e valores das paróquias da arquidiocese

```python
url_principal = "http://arquisp.org.br/buscar-paroquias?field_lugar_paroquia_target_id=19449&combine=&field_lugar_endereco_thoroughfare=&field_lugar_endereco_dependent_locality="
req = urllib.request.Request(
        url_principal, 
        data=None, 
        headers={
        'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'
    }
)
f = urllib.request.urlopen(req)
content = bf4(f.read())
```

```python
paroquia = select(content,'#edit-field-lugar-paroquia-target-id')
paroquia = str(paroquia)
```

```python
parI = find_substring("<option",paroquia)
parF = find_substring("</option>",paroquia)
valI = list(np.array(parI)+15)
valF = list(np.array(valI)+5) 
```

```python
valor = []
nome = []
x=1
inicio = list(np.array(parI)+24)
fim = list(np.array(parF)-1)
while x < len(parI):
    valor.append(paroquia[valI[x]:valF[x]])
    nome.append(paroquia[inicio[x]:fim[x]])
    x+=1
valor[34]='19449'
len(valor)
```

# Pegando dentro de cada paróquia as igrejas e endereços

```python
listao =[]
for u in valor:   
    url_principal = "http://arquisp.org.br/buscar-paroquias?field_lugar_paroquia_target_id="
    url_secundario = u
    url_terceiro = "&combine=&field_lugar_endereco_thoroughfare=&field_lugar_endereco_dependent_locality="
    url = url_principal+url_secundario+url_terceiro
    req = urllib.request.Request(
            url, 
            data=None, 
            headers={
            'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'
        }
    )
    f = urllib.request.urlopen(req)
    content1 = bf4(f.read())
    igrejas = select(content1,'.view-content')
    igrejas =str(igrejas)
    igrI = find_substring("<span class=",igrejas)
    igrF = find_substring("</span>",igrejas)
    endI = find_substring('<div class="views-field views-field-field-lugar-endereco-premise">',igrejas)
    endF = find_substring('</div>',igrejas)
    endI
    endFn =[]
    j=0
    for i in range(len(endI)):
        while j < len(endF):
            if endI[i]<endF[j]:
                endFn.append(endF[j])
                j=len(endF)
            else:
                j+=1
        j=0
    igr_nomes = []
    endereco = []
    x=0
    while x <len(igrI):
        igr_nomes.append(igrejas[igrI[x]:igrF[x]])
        endereco.append(igrejas[endI[x]:endFn[x]])
        x+=1
    adress = []
    church = []
    i=0
    while i< len(endereco):
        adress.append(endereco[i][94:])
        i+=1
    for k in range(len(igr_nomes)):
        iniciando = find_substring("\">",igr_nomes[k])
        finalizando = find_substring("</a>",igr_nomes[k])
        iniciando = list(np.array(iniciando)+2)
        church.append(igr_nomes[k][iniciando[1]:finalizando[0]])
    listao.append([church,adress])
    print("Já foi a página : {}".format(len(listao)))
    sleep(random.uniform(0.5,1.5))    
```

```python
data = listao
```

```python
igreja = []
local =[]
for s in range(1,len(data)):
    data[0][0] += data[s][0]
    data[0][1] += data[s][1]
igreja.append(data[0][0])
local.append(data[0][1])
```

```python
nomes = []
enderecos =[]
while y < len(igreja[0]):
    nomes.append(igreja[0][y])
    enderecos.append(local[0][y])
    y+=1
```

```python
dados = pd.DataFrame({'Nomes':nomes,'Endereco':enderecos})
dados = dados[dados.Nomes.str.contains('Matriz|Área')].reset_index(drop=True)
dados
```
